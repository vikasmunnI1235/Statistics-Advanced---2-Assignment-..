Question 1: What is hypothesis testing in statistics?

ANS
:- Hypothesis testing is like a court trial:

H‚ÇÄ = "Defendant is innocent."

Evidence (data) is presented.

If evidence is strong enough (p-value small), reject H‚ÇÄ and conclude "defendant is guilty."

hypothesis?

H‚ÇÄ = "Nothing is happening."

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

ANS :- The significance level (Œ±) is the cut-off line that helps us decide:

Is the observed result just due to random chance (retain H‚ÇÄ)?

Or is it strong enough evidence to reject H‚ÇÄ and support H‚ÇÅ?

Question 4: What are Type I and Type II errors? Give examples of each.

ANS
| Error Type      | What happens?                                       | Real-world analogy                  |
| --------------- | --------------------------------------------------- | ----------------------------------- |
| **Type I (Œ±)**  | Reject H‚ÇÄ when H‚ÇÄ is true (false positive)          | Declaring an innocent person guilty |
| **Type II (Œ≤)** | Fail to reject H‚ÇÄ when H‚ÇÄ is false (false negative) | Letting a guilty person go free     |


Type I = false alarm.

Type II = missed detection.

Question 6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

ANS

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10     # number of trials
p = 0.5    # probability of success
size = 1000  # number of samples

# Generate random samples from a binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.show()
```

Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

ANS

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Sample statistics
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation

# Z-statistic calculation
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Evidence suggests the mean is different from", mu_0)
else:
    print("Fail to Reject H0: Not enough evidence to say the mean is different from", mu_0)
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

ANS

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # population mean and std dev
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

# Step 3: Compute 95% confidence interval
confidence = 0.95
z_critical = stats.norm.ppf(1 - (1-confidence)/2)  # two-tailed
margin_of_error = z_critical * se
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Step 4: Plot histogram of data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='blue', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Sample with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation

    # Compute Z-scores
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores)

# Example dataset
data = [10, 12, 13, 15, 16, 18, 20, 22, 23, 25]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Original Data:", data)
print("Z-scores:", z_scores)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score (Standard Deviations)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
